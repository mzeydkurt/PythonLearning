```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Helikopter_Verileri.xlsx')
```

```python
df
```

```python
# @title Azami Kalkış Ağırlığı (kg)

from matplotlib import pyplot as plt
df['Azami Kalkış Ağırlığı (kg)'].plot(kind='line', figsize=(8, 4), title='Azami Kalkış Ağırlığı (kg)')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# @title Havada Kalma Süresi (saat)

from matplotlib import pyplot as plt
df['Havada Kalma Süresi (saat)'].plot(kind='hist', bins=20, title='Havada Kalma Süresi (saat)')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Azami Kalkış Ağırlığı (kg) vs Faydalı Yük Kapasitesi (kg)

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Azami Kalkış Ağırlığı (kg)', y='Faydalı Yük Kapasitesi (kg)', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Motor Modeli

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Motor Modeli').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Havada Kalma Süresi (saat) vs Menzil - Kontrol YarıÇapı (km)

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Havada Kalma Süresi (saat)', y='Menzil - Kontrol YarıÇapı (km)', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```
